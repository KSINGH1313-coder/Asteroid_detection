```python
df=df.drop(['orbiting_body'],axis=1)
```

```python
df['is_hazardous']=df['is_hazardous'].apply(lambda x:1 if x==True else 0)
```

```python
df['id'] = df['name'].str.extract(r'(\d+)')
```

```python
df.head()
```

```python
df.drop('name',axis=1,inplace=True)
```

```python
df.isnull().sum()
```

```python
df['absolute_magnitude']=df['absolute_magnitude'].fillna(df['absolute_magnitude'].mean())
df['estimated_diameter_max']=df['estimated_diameter_max'].fillna(df['estimated_diameter_max'].mean())
df['estimated_diameter_min']=df['estimated_diameter_min'].fillna(df['estimated_diameter_min'].mean())
```

```python
x=df.drop('is_hazardous',axis=1)
y=df['is_hazardous']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
dict={}
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train,y_train)
y_pred=lr.predict(x_test)
dict['LogisticRegression']=accuracy_score(y_test,y_pred)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
dt.fit(x_train,y_train)
y_pred=dt.predict(x_test)
dict['DecisionTreeClassifier']=accuracy_score(y_test,y_pred)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
rf.fit(x_train,y_train)
y_pred=rf.predict(x_test)
dict['RandomForestClassifier']=accuracy_score(y_test,y_pred)
print(accuracy_score(y_test,y_pred))
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()
knn.fit(x_train,y_train)
y_pred=knn.predict(x_test)
dict['KNeighborsClassifier']=accuracy_score(y_test,y_pred)
```

```python
from sklearn.naive_bayes import GaussianNB
nb=GaussianNB()
nb.fit(x_train,y_train)
y_pred=nb.predict(x_test)
dict['GaussianNB']=accuracy_score(y_test,y_pred)
```

```python
plt.bar(dict.keys(),dict.values())
plt.xticks(rotation=90)
plt.show()
```
